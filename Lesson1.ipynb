```python
# 1. 
# Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, 
# сохранить JSON-вывод в файле *.json.
```

```python
import requests
from pprint import pprint
```

```python
url = 'https://api.github.com/users/ksenyakuz/repos'

response = requests.get(url=url)
print(response.status_code)
pprint(response.json())
```

```python
repo = []

for i in response.json():
    repo.append(i['name'])
    
print(f'Список репозиториев пользователя:')
print(repo)
```

```python
import json

with open('repos.txt', 'w') as outfile:
    json.dump(repo, outfile)
```

```python
#2. 
# Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search
# (а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. 
# Для каждой категории рассчитать среднюю минимальную стоимость заказа.
# (б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов
```

```python
params = {
    'latitude' : '59.989931',
    'longitude' : '30.241865',
    'query' : 'панна котта'
}

url = 'https://api.delivery-club.ru/api1.2/vendors/search'

response = requests.get(url=url, params=params)

print(response.status_code)
pprint(response.json())
```

```python
vendors = response.json()['vendors']
```

```python
names = []
category = []
delivery_price = []
min_order_price = []

for vendor in vendors:
    names.append(vendor['name'])
    category.append(vendor['categoryId'])
    delivery_price.append(vendor['delivery']['price']['value'])
    min_order_price.append(vendor['delivery']['minOrderPrice']['value'])

print(names)
print(category)
print(delivery_price)
print(min_order_price)
```
